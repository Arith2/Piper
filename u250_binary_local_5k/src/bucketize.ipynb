```python
import torch
```

```python
input_size = 16000000

boundary_size = 4096
```

```python
import numpy as np

rng = np.random.default_rng(20230)

bucketize_boundary = np.arange(boundary_size) * 10
bucketize_input = rng.integers(low=0, high=boundary_size*10, size=input_size)
```

```python
bucketize_boundary_tensor = torch.from_numpy(bucketize_boundary)
bucketize_input_tensor = torch.from_numpy(bucketize_input)
```

```python
import time

t1 = time.perf_counter()

bucketize_result = torch.bucketize(bucketize_input_tensor, bucketize_boundary_tensor)

t2 = time.perf_counter()

print("Execution time: %s ms", (t2-t1)*1000)
```
